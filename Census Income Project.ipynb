# Census Income Project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("census-income.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.replace("?",np.nan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.columns
```

```python
for col in df.columns:
    if df[col].dtype=="object":
        df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
df.info()
```

# Model Building

```python
# Split Data into Dependent and Independent
x=df.iloc[:,:-1]
```

```python
#independent
x
```

```python
y=df["annual_income"]
```

```python
#dependent
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=56)
```

# Testing Predictions with different models

# 1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
```

```python
lr_model=LogisticRegression()
```

```python
lr_model
```

```python
#Training the model
lr_model.fit(x_train,y_train)
```

```python
#Testing the model
lr_pred=lr_model.predict(x_test)
lr_pred
```

```python
y_test
```

```python
from sklearn.metrics import *
```

```python
lr_acc=accuracy_score(y_test,lr_pred)
lr_acc
```

# 2. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model=DecisionTreeClassifier()
```

```python
#Training the model
dt_model.fit(x_train,y_train)
```

```python
#Testing the model
dt_pred=dt_model.predict(x_test)
dt_pred
```

```python
y_test
```

```python
dt_acc=accuracy_score(y_test,dt_pred)
dt_acc
```

# 3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier()
```

```python
#Training the model
rf_model.fit(x_train,y_train)
```

```python
#Testing the model
rf_pred=rf_model.predict(x_test)
rf_pred
```

```python
y_test
```

```python
rf_acc=accuracy_score(y_test,rf_pred)
```

```python
rf_acc
```

```python
print("Accuracy of Logistic Regression",lr_acc)
print("Accuracy of Decision Tree",dt_acc)
print("Accuracy of Random Forest",rf_acc)
```

```python
# Since random forest gives more acccuracy we will further optimize it.
```

```python
trees=[i for i in range(50,800,50)]
```

```python
trees
```

```python

for N in trees:
    rf_model=RandomForestClassifier(n_estimators=N,random_state=56)
    rf_model.fit(x_train,y_train)
    rf_pred=rf_model.predict(x_test)
    acc=accuracy_score(y_test,rf_pred)
    print("Random forest with",N," number of trees having accuracy",acc)
```
